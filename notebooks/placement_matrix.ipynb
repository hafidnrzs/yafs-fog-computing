```python
import json
from collections import defaultdict

# json_data = {
#   "initialAllocation": [
#     {"module_name": "0_0", "app": "0", "id_resource": 36},
#     {"module_name": "0_0", "app": "0", "id_resource": 100},
#     {"module_name": "0_1", "app": "0", "id_resource": 36},
#     {"module_name": "1_3", "app": "1", "id_resource": 26},
#     {"module_name": "1_3", "app": "1", "id_resource": 69},
#     {"module_name": "1_3", "app": "1", "id_resource": 74}
#   ]
# }
json_data = json.load(open('allocDefinition.json'))

# Membuat placementMatrix dari data JSON
# Menggunakan defaultdict(list) agar lebih mudah menambahkan ID resource
placement_matrix = defaultdict(list)

for allocation in json_data["initialAllocation"]:
    module = allocation["module_name"]
    resource_id = allocation["id_resource"]
    # Menambahkan resource_id ke daftar untuk modul yang bersangkutan
    if resource_id not in placement_matrix[module]:
         placement_matrix[module].append(resource_id)

# Mencetak hasilnya agar terlihat jelas
# Mengubah kembali ke dict biasa untuk tampilan yang bersih
print(json.dumps(dict(placement_matrix), indent=2))

# export placement_matrix ke json
with open('placement_matrix.json', 'w') as f:
    json.dump(dict(placement_matrix), f, indent=2)
```
