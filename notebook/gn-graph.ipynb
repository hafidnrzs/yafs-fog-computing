# GN Graph

```python
import networkx as nx
import matplotlib.pyplot as plt

# Jumlah node (n) untuk dieksplor
node_values = [2, 5, 10]
seed_value = 42

# Atur plot 2 baris, 3 kolom (baris pertama: original, baris kedua: reverse)
fig, axes = plt.subplots(2, len(node_values), figsize=(5 * len(node_values), 10))

print("--- Graf Growing Network (GN) dan Setelah Di-Reverse ---")

for i, node_val in enumerate(node_values):
    # Buat graf GN
    G_gn = nx.gn_graph(n=node_val, seed=seed_value)
    G_rev = G_gn.reverse(copy=True)  # Safe reverse

    print(f"\n--- GN Graph with n={node_val} ---")
    print(f"Original edges: {list(G_gn.edges)}")
    print(f"Reversed edges: {list(G_rev.edges)}")

    pos = nx.spring_layout(G_gn, seed=seed_value)

    # Baris 0: original
    ax_orig = axes[0][i]
    nx.draw(G_gn, pos, ax=ax_orig, with_labels=True,
            node_color='skyblue', node_size=800,
            edge_color='gray', arrows=True)
    ax_orig.set_title(f"Original GN (n={node_val})")

    # Baris 1: reversed
    ax_rev = axes[1][i]
    nx.draw(G_rev, pos, ax=ax_rev, with_labels=True,
            node_color='lightgreen', node_size=800,
            edge_color='red', arrows=True)
    ax_rev.set_title(f"Reversed GN (n={node_val})")

# Final layout fix
plt.tight_layout()
plt.show()
```
